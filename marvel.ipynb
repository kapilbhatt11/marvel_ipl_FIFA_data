```python
marvel['Alignment'].value_counts()
```

```python
good = marvel[marvel["Alignment"]=="good"]
```

```python
good.head()
```

```python
good.sort_values(by='Speed',ascending=False)
```

```python
good.sort_values(by=['Speed'],ascending=False).head()
```

```python
good.sort_values(by=['Power'],ascending=False).head()
```

```python
good_max_power = good[good['Power']==100]
```

```python
good_max_power.shape
```

```python
good_max_power
```

```python
good_max_power_total = good.sort_values(by=['Total'],ascending=False)
```

```python
good_max_power_total.head()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(good_max_power_total['Name'])[0:5],list(good_max_power_total['Total'])[0:5],color ='g')
plt.show()
```
